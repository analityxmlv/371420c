## 💾 Los datos
Tenemos acceso a diez años de información de pacientes ([source](https://archive.ics.uci.edu/ml/datasets/Diabetes+130-US+hospitals+for+years+1999-2008)):

#### Información n el archivo
- "age" - rango de edad del paciente
- "time_in_hospital" - dias (de 1 to 14)
- "n_procedures" - número de procedimientos por los que paso durante su estancia en el hospital
- "n_lab_procedures" - número de procedimientos de laboratorio hechos durante su tiempo en el hospital
- "n_medications" - número de medicamentos administrados durante su estancia en el hospital
- "n_outpatient" - número de visitas de pacientes externos durante el año antes de ser internado
- "n_inpatient" - número de visitas de pacientes internados antes de quedarse en el hospital
- "n_emergency" - número de visitas a la sala de emergencias en el año antes de ser hospitalizado
- "medical_specialty" - la especialidad del médico de admisión
- "diag_1" - diagnóstico primario (Circulatory, Respiratory, Digestive, etc.)
- "diag_2" - diagnóstico secundario
- "diag_3" - diagnóstico secundario adicional 
- "glucose_test" - sean los resultados de suero de glucosa high (> 200), normal, or not performed
- "A1Ctest" - sean los niveles de A1C resultantes del paciente, high (> 7%), normal, or not performed
- "change" - a existido un cambio en la medicación de diabetes ('yes' or 'no')
- "diabetes_med" - se ha prescrito medicación para diabetes ('yes' or 'no')
- "readmitted" - el paciente a sido readmitido en el hospital ('yes' or 'no')
